## AI Gym installation
We will be using a library called Gymnasium to run some simple reinforcemnet learning environments. If you want to run these locally on your machine you will need to install it, otherwise you can just run on google colab.

The documention and install instructions can be found here: https://github.com/Farama-Foundation/Gymnasium


```python
# run this cell to check your installation. 
# A simple text based output should appear 
import gym
import numpy as np
env = gym.make('CliffWalking-v0')
env.render_mode = 'human'
env.reset()

for _ in range(100):
    action = np.random.randint(0,4)  # User-defined policy function
    next_state, reward, done, info = env.step(action)
    env.render(mode='human')

    if done:
        break
env.close()
```
